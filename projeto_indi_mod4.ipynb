#Criação da Tabela

```python
#Import das bibliotecas pandas e numpy
import pandas as pd
import numpy as np

#Criação da tabela que é mostrado no slide com array
dados = {'Dias': ['Segunda','Terça', 'Quarta', 'Quinta', 'Sexta','Sábado','Domingo'],
         'Limpeza':[100,0,100,0,100,100,0],
         'Comida':[221.60,375.31,412.00,495.20,411.53,245.00,164.00],
         'Transporte':[150,100,125,300,275,525,75],
         'Outros':[0,0,2310,500,0,0,820]}
df = pd.DataFrame(dados,index = [170423,180423,190423,200423,210423,220423,230423])
df
```

```python
#Inclusão dos ganhos na tabela

df['Ganhos'] = [2200,2420.50,3391,5322,4898.50,4200,3893]
df
```

#Gerando Relatorio

**Gere um relatório de uma semana para mostrar ao dono da loja como a análise dos
dados pode ser útil e incluir no relatório outros itens:**

1. A subtração de impostos dos ganhos diários, que nesta semana foi de 7%;
2. A soma total dos ganhos;
3. A média semanal dos ganhos;
4. A soma total das despesas por categoria;
5. A média semanal de todas as despesas;

os resultados bem apresentados.


# 1.Subtração  de impostos dos ganhos diários

```python
#Criando novas tabelas para relalziar o calculo
df['Impostos'] = (df['Ganhos'] * 7)/100
df['Ganhos s/Impostos'] = df['Ganhos'] - df['Impostos']
df


#Ou realizar a mesma operação sem precisar criar novas tabelas
#calcular_imposto = lambda x: (x - (x*7)/100)
#print(calcular_imposto(df['Ganhos']))
```

#2.Soma total dos ganhos


```python
#Somando todaa coluna Ganhos utilziando a função SUM
soma_ganhos = df['Ganhos'].sum()
soma_ganhos
```

# 3.Média Semanal dos ganhos

```python
#Pegando a média da coluna Ganhos usando a função MEAN e usando FORMAT para uma melhor visualização.

media = df['Ganhos'].mean()
print ('{:.2f}'.format(media))
```

# 4.Soma Total das despesas por categoria

```python
#Foi realizado um print usando a função SUM para somar os valores da linha e dar o total da coluna de cada 
print('A soma total da despesa com limpeza é de : ',df['Limpeza'].sum(),'\nA soma total da despesa com comida é de : ',df['Comida'].sum(),'\nA soma total da despesa com transporte é de : ',df['Transporte'].sum(),'\nA soma total da despesa com outros é de : ',df['Outros'].sum())
```

# 5.Media semanal de todas as despesas

```python
soma_despesas = (df['Limpeza'].sum() + df['Comida'].sum() + df['Transporte'].sum() + df['Outros'].sum())/ 4
soma_despesas
print('O valor da média semanal das despesas foi de : ', '{:.2f}'.format(soma_despesas))
```

#6.Dia mais Lucrativo e o lucro total da semana.


```python
#Lucro total de cada dia
#Criei uma nova tabela chamada Despesas, onde ele soma todos os valores das despesas Limpeza,Comida, Transporte e Outros.
#Depois criei uma nova tabela chamada Lucro, onde ela pega os Ganhos já diminuidos sem impostos e diminuidos pela despesas.
#Assim, na tabela Lucro fica aparecendo o lucro de cada dia.
df['Despesas'] = df[['Limpeza','Comida','Transporte','Outros']].sum(axis = 1)
df['Lucro'] = df['Ganhos s/Impostos'] - df['Despesas']
df['Lucro'].max()
df
```

```python
#Dia mais lucrativo
#Fiz uma váriavel selecionando as colunas Dias e Lucro, depois utilizei a função SORT_VALUES e ordenei pelo maior Lucro.

Dia_mais_Lucrativo = df[['Dias','Lucro']]
Dia_mais_Lucrativo.sort_values(by = 'Lucro',ascending=False ).round(2)
```

```python
#Lucro total da semana
soma_lucro = df['Lucro'].sum()
print('O Lucro total da semana foi de: ', soma_lucro)
```
